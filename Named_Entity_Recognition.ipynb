```python
import pandas as pd
import spacy
import stanza


df1 = pd.read_csv('/content/cleaned_computer_sc.csv')
df2 = pd.read_csv('/content/cleaned_sculptors.csv')

# Concatenate the dataframes
df = pd.concat([df1, df2])

# Print column names to verify the actual column name
print("Column names in the DataFrame:", df.columns)

# Ensure the correct column name is used
correct_column_name = 'Content'  # Adjust based on actual column name

# Function to process a document chunk
def process_chunk(text_chunk, spacy_model, stanza_model):
    # Process the text chunk with Spacy
    spacy_doc = spacy_model(text_chunk)
    print("Spacy Entities:")
    for ent in spacy_doc.ents:
        print(f"Entity: {ent.text}, Type: {ent.label_}")

    # Process the text chunk with Stanza
    stanza_doc = stanza_model(text_chunk)
    print("Stanza Entities:")
    for sentence in stanza_doc.sentences:
        for entity in sentence.ents:
            print(f"Entity: {entity.text}, Type: {entity.type}")

# Process the text in smaller chunks
chunk_size = 100  # Adjust this value based on your available memory
for i in range(0, len(df), chunk_size):
    chunk = df[correct_column_name].iloc[i: i + chunk_size].tolist()
    text_chunk = ' '.join(chunk)
    process_chunk(text_chunk, spacy_model, stanza_model)
```

```python
# prompt: Write a function that will process one document (i.e. a biography)
# and return the set of named entities (NE) detected within it. Do
# this for both Spacy and Stanza. Store your results where (e.g.
# Pandas dataframe/json file) you can easily retrieve them to do
# comparisons.this is the data df1 = pd.read_csv('/content/cleaned_computer_sc.csv')
# df2 = pd.read_csv('/content/cleaned_sculptors.csv')
# df = pd


def extract_entities_spacy(text):
    doc = nlp_spacy(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

def process_document_spacy(text):
    spacy_ents = extract_entities_spacy(text)
    return {'text': text, 'spacy': spacy_ents}

def store_results_spacy(results):
    df = pd.DataFrame(results)
    df.to_json('named_entities_spacy.json', orient='records')
    return df

# Load the data
df1 = pd.read_csv('/content/cleaned_computer_sc.csv')
df2 = pd.read_csv('/content/cleaned_sculptors.csv')
df = pd.concat([df1, df2])

# Extract the text column
texts = df['Content'].tolist()

# Process each document and store results
results_spacy = [process_document_spacy(text) for text in texts]

# Store the results in a DataFrame and save to JSON
df_results_spacy = store_results_spacy(results_spacy)

# Print a sample of the results
print(df_results_spacy.head())
```

```python
def extract_entities_stanza(text):
    doc = nlp_stanza(text)
    entities = [(ent.text, ent.type) for ent in doc.ents]
    return entities

def process_document_stanza(text):
    stanza_ents = extract_entities_stanza(text)
    return {'text': text, 'stanza': stanza_ents}

def store_results_stanza(results):
    df = pd.DataFrame(results)
    df.to_json('named_entities_stanza.json', orient='records')
    return df

# Load the data
df1 = pd.read_csv('/content/cleaned_computer_sc.csv')
df2 = pd.read_csv('/content/cleaned_sculptors.csv')
df = pd.concat([df1, df2])

# Extract the text column
texts = df['Content'].tolist()

# Process each document and store results
results_stanza = [process_document_stanza(text) for text in texts]

# Store the results in a DataFrame and save to JSON
df_results_stanza = store_results_stanza(results_stanza)

# Print a sample of the results
print(df_results_stanza.head())
```

```python
import pandas as pd
def ner_comparison(doc, spacy_model, stanza_model):
    # Tokenize the document using both Spacy and Stanza
    spacy_doc = spacy_model(doc)
    stanza_doc = stanza_model(doc)

    spacy_entities = [(ent.text, ent.label_) for ent in spacy_doc.ents]
    stanza_entities = [(ent.text, ent.type) for sent in stanza_doc.sentences for ent in sent.ents]

    spacy_spans = [(ent.start_char, ent.end_char) for ent in spacy_doc.ents]
    stanza_spans = [(ent.start_char, ent.end_char) for sent in stanza_doc.sentences for ent in sent.ents]

    # Initialize counters
    full_agreement = 0
    partial_agreement = 0
    spacy_only = 0
    stanza_only = 0
    agreement_types = 0

    # Check for overlaps and agreements
    for spacy_span in spacy_spans:
        spacy_span_text = doc[spacy_span[0]:spacy_span[1]]
        for stanza_span in stanza_spans:
            stanza_span_text = doc[stanza_span[0]:stanza_span[1]]
            if spacy_span == stanza_span:
                full_agreement += 1
                if spacy_span_text == stanza_span_text:
                    agreement_types += 1
            elif spacy_span[0] < stanza_span[1] and spacy_span[1] > stanza_span[0]:
                partial_agreement += 1

    for spacy_span in spacy_spans:
        if spacy_span not in stanza_spans:
            spacy_only += 1

    for stanza_span in stanza_spans:
        if stanza_span not in spacy_spans:
            stanza_only += 1

    return {
        'full_agreement': full_agreement,
        'partial_agreement': partial_agreement,
        'spacy_only': spacy_only,
        'stanza_only': stanza_only,
        'agreement_types': agreement_types
    }

# Load the dataset
data = pd.read_csv("/content/cleaned_computer_sc.csv")

# Initialize Spacy and Stanza models
spacy_model = spacy.load("en_core_web_sm") # Load the Spacy model after importing the library
stanza_model = stanza.Pipeline('en')

# Process each document in the dataset
statistics = []
for doc in data['content']:
    stats = ner_comparison(doc, spacy_model, stanza_model)
    statistics.append(stats)

# Visualize the statistics
# You can use your preferred visualization library like matplotlib or seaborn to plot the statistics.
# For simplicity, I'm just printing the statistics here.
for i, stats in enumerate(statistics):
    print(f"Document {i+1}:")
    print("Full Agreement:", stats['full_agreement'])
    print("Partial Agreement:", stats['partial_agreement'])
    print("Spacy Only:", stats['spacy_only'])
    print("Stanza Only:", stats['stanza_only'])
    print("Agreement in NE Type:", stats['agreement_types'])
    print()
```
