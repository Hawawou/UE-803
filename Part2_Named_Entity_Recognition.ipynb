```python
import spacy

def extract_entities_spacy(text):
    # Load the SpaCy model
    nlp = spacy.load("en_core_web_sm")

    # Process the document
    doc = nlp(text)

    # Extract named entities
    entities = set()
    for ent in doc.ents:
        entities.add((ent.text, ent.label_))

    return entities

# Example usage
text = "Apple is looking at buying U.K. startup for $1 billion."
entities = extract_entities_spacy(text)
print(entities)
```

```python
import stanza

def extract_entities_stanza(text):
    # Initialize the pipeline
    nlp = stanza.Pipeline('en')

    # Process the document
    doc = nlp(text)

    # Extract named entities
    entities = set()
    for sentence in doc.sentences:
        for ent in sentence.ents:
            entities.add((ent.text, ent.type))

    return entities

# Example usage
text = "Apple is looking at buying U.K. startup for $1 billion."
entities = extract_entities_stanza(text)
print(entities)
```

```python
import pandas as pd
df1 = pd.read_csv('/content/biography_cs.csv')
df2 = pd.read_csv('/content/biography_sculptor.csv')
df = pd.concat([df1, df2])
df.head()
```

```python
NE_spacy= []
NE_stanza=[]
for text in df['Content']:
    NE_spacy.append(extract_entities_spacy(text))
    NE_stanza.append(extract_entities_stanza(text))
df['NE_spacy'] = NE_spacy
df['NE_stanza'] = NE_stanza
df.head()
```
