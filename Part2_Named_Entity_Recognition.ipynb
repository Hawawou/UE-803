## Spacy

```python
import spacy

def extract_entities_spacy(text):
    # Load the SpaCy model
    nlp = spacy.load("en_core_web_sm")

    # Process the document
    doc = nlp(text)

    # Extract named entities
    entities = set()
    for ent in doc.ents:
        entities.add((ent.text, ent.label_))

    return entities

# Example usage
# text = "Apple is looking at buying U.K. startup for $1 billion."
# entities = extract_entities_spacy(text)
# print(entities)
```

## Stanza

```python
import stanza

def extract_entities_stanza(text):
    # Initialize the pipeline
    nlp = stanza.Pipeline('en')

    # Process the document
    doc = nlp(text)

    # Extract named entities
    entities = set()
    for sentence in doc.sentences:
        for ent in sentence.ents:
            entities.add((ent.text, ent.type))

    return entities

# Example usage
text = "Apple is looking at buying U.K. startup for $1 billion."
entities = extract_entities_stanza(text)
print(entities)
```

Basic statistics for Name Entities Spacy vs. Stanza

--avg/min/max/number of NE per category per package

--avg/min/max/number of words in each NE per category per package

-- visualizations to compare the above 2 statistics per category per package

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



# Calculate the number of named entities per text
df['num_NE_spacy'] = df['NE_spacy'].apply(len)
df['num_NE_stanza'] = df['NE_stanza'].apply(len)

# Calculate the number of words in each named entity and then the average for each text
df['avg_words_NE_spacy'] = df['NE_spacy'].apply(lambda x: np.mean([len(entity.split()) for entity in x]) if x else 0)
df['avg_words_NE_stanza'] = df['NE_stanza'].apply(lambda x: np.mean([len(entity.split()) for entity in x]) if x else 0)

# Group by Category and calculate statistics
stats_spacy = df.groupby('Category')['num_NE_spacy'].agg(['mean', 'min', 'max', 'count'])
stats_stanza = df.groupby('Category')['num_NE_stanza'].agg(['mean', 'min', 'max', 'count'])

words_stats_spacy = df.groupby('Category')['avg_words_NE_spacy'].agg(['mean', 'min', 'max'])
words_stats_stanza = df.groupby('Category')['avg_words_NE_stanza'].agg(['mean', 'min', 'max'])

# Print statistics
print("Named Entity Statistics for spaCy:")
print(stats_spacy)
print("\nNamed Entity Statistics for Stanza:")
print(stats_stanza)

print("\nAverage Number of Words in Named Entities for spaCy:")
print(words_stats_spacy)
print("\nAverage Number of Words in Named Entities for Stanza:")
print(words_stats_stanza)

# Visualization
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Number of NEs
axs[0, 0].bar(stats_spacy.index, stats_spacy['mean'], color='b', alpha=0.6, label='spaCy')
axs[0, 0].bar(stats_stanza.index, stats_stanza['mean'], color='r', alpha=0.6, label='Stanza')
axs[0, 0].set_title('Average Number of NEs')
axs[0, 0].legend()

axs[0, 1].bar(stats_spacy.index, stats_spacy['max'], color='b', alpha=0.6, label='spaCy')
axs[0, 1].bar(stats_stanza.index, stats_stanza['max'], color='r', alpha=0.6, label='Stanza')
axs[0, 1].set_title('Maximum Number of NEs')
axs[0, 1].legend()

# Average number of words in NEs
axs[1, 0].bar(words_stats_spacy.index, words_stats_spacy['mean'], color='b', alpha=0.6, label='spaCy')
axs[1, 0].bar(words_stats_stanza.index, words_stats_stanza['mean'], color='r', alpha=0.6, label='Stanza')
axs[1, 0].set_title('Average Number of Words in NEs')
axs[1, 0].legend()

axs[1, 1].bar(words_stats_spacy.index, words_stats_spacy['max'], color='b', alpha=0.6, label='spaCy')
axs[1, 1].bar(words_stats_stanza.index, words_stats_stanza['max'], color='r', alpha=0.6, label='Stanza')
axs[1, 1].set_title('Maximum Number of Words in NEs')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

## Verification against KG

```python
rdf_data = pd.read_csv('rdf_cleaned.csv')
rdf_data.head()
```

```python
kg_data = rdf_data.groupby('Subject')['Object'].apply(list).to_dict()
kg_data
```

```python
import re
```

```python
def evaluate_ne_to_kg(kg_data, data):
    results = {}
    for index, row in data.iterrows():
        person = row['Name']
        ne_list = row['NE_spacy']
       
        kg_entities = kg_data.get(person, [])

        # Initialize counter
        s_total = len(ne_list)
        s_matched = 0
        
        # Check if any named entity matches a KG entity
        def match_ne_to_kg(ne, kg_entities):
            ne_text, ne_label = ne
            for kg_entity in kg_entities:
                
                # Use re to find matches considering potential variations in text
                pattern = re.escape(ne_text)
                if re.search(pattern, kg_entity, re.IGNORECASE):
                    return True
            return False
        
        # Evaluate NE prediction for spacy
        for ne in ne_list:
            if match_ne_to_kg(ne, kg_entities):
                s_matched += 1

        # Ratio
        spacy_ratio = s_matched / s_total if s_total > 0 else 0

        results[person] = {'spacy': spacy_ratio}

    return results

```
