```python
from spacy.tokenizer import Tokenizer
from spacy.lang.en import English
import pandas as pd
```

```python
sc = pd.read_csv('cleaned_sculptors.csv')
cs = pd.read_csv('cleaned_computer_sc.csv')
df = pd.concat([sc, cs], ignore_index=True)
df.tail()
```

```python
# Create a tokenizer with default settings for English
nlp = English()
tokenizer = nlp.tokenizer
```

```python
df_tokenised = df['Content'].apply(lambda x: tokenizer(x))
df_tokenised.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk import word_tokenize
```

```python
vectorizer = TfidfVectorizer(max_features=500,
                             use_idf=True,
                             stop_words='english',
                             tokenizer=nltk.word_tokenize)

matrix = vectorizer.fit_transform(df['Content'])
```

```python
features = vectorizer.get_feature_names_out()
features
```

```python
print(matrix)
```

### K-means clustering

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=16, init='k-means++', max_iter=300, n_init=5, verbose=0, random_state=2425)
km.fit(matrix)
```

```python
predicted_labels = km.labels_
print(predicted_labels)
predicted_labels = predicted_labels.tolist()
```
